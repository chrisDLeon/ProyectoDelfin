```python
n = df_2011.columns.values.tolist()
for i in n:
    if df_2011[i].value_counts().get("T") != None:
        if df_2011[i].mode().tolist()[0] == 'T':
            m = df_2011.drop(df_2011[(df_2011[i] == 'T')].index)
            m= m[i].mode().tolist()
            df_2011[i].replace({'T':m[0]}, inplace=True)
        else:
            df_2011[i].replace({'T':df_2011[i].mode().tolist()[0]}, inplace=True)
df_2011.head()
```
